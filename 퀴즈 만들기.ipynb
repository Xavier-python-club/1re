\\$

-------------------

$\\
\\$

```
q1= input("한글을 만든 왕은?: ")
a1= '세종대왕'

q2= input("캐나다의 수도는?: ")
a2='오타와'

q3= input("2x+4 = 3x+10 일 때, x의 값은?: ")
a3='-6'

q4= input("원소기호가 'C'인 원자는?: ")
a4= '탄소'

q5= input("태양계에서 가장 큰 행성은?: ")
a5= '목성'
```

`input()`을 사용하여 사용자의 커맨드창에 문제들이 나타나게 하고, 사용자가 답을 입력할 수 있게 해줍니다.

$\\$

```
res = 0
```
점수가 될 res는 0으로 설정해줍니다. 

$\\$

```
if q1 == a1:
    res = res+20

if q2 == a2:
    res = res+20
    
   ...
   
```
`if`문을 사용하여 만약 사용자가 입력한 답이 문제에 답과 일치한다면 점수에 20을 추가합니다. 같은 방식으로 문제의 갯수만큼 반복합니다.

$\\$

```
print('score:'+str(res))
```
마지막으로 score문구와 함께 점수를 출력해줍니다. res는 정수값이므로 문자열 형식으로 변환해줍니다.

$\\$

```
if res == 100:
    print('congratulations!')
```
사용자의 점수가 100점이면 'congratulations!'라는 문구를 함께 띄웁니다. 

$\\$

다음은 프로그램을 실행했을 때의 결과입니다. 

````python
q1= input("한글을 만든 왕은?: ")
a1= '세종대왕'

q2= input("캐나다의 수도는?: ")
a2='오타와'

q3= input("2x+4 = 3x+10 일 때, x의 값은?: ")
a3='-6'

q4= input("원소기호가 'C'인 원자는?: ")
a4= '탄소'

q5= input("태양계에서 가장 큰 행성은?: ")
a5= '목성'

res = 0

if q1 == a1:
    res = res+20

if q2 == a2:
    res = res+20

if q3 == a3:
    res = res+20
    
if q4 == a4:
    res = res+20
    
if q5 == a5:
    res = res+20

print('score:'+str(res))

if res == 100:
    print('congratulations!')
````
