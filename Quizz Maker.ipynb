By using 'input()', the questions will show up on the command screen, enableing the user to respond.

We will set res to 0, which will be the initial score.


    if q1 == a1:
        res = res + 20

    if q2 == a2:
        res = res + 20

        ...

Here, we will use the 'if' statement, so that if the answer of the user corresponds to the one of the question, 20 points will be added to the score. 
Repeat for the amount of questions.


Make sure not confuse = and == , which are two completely different conditions.

a = b means b is attributed to a, so the original value of a is changed to the value of b.
a == b means comparing if the two values are equal, which is the same as the normal = we use in common life.


    print('score:'+str(res))

Lastly, we will print the final score with 'score:'. Since res is an integer, we will convert it into a string.


    if res == 100:
        print('congratulations!')
    else:
        print('better luck next time ;)')

The phrase 'congratulations!' will be shown if the final score is equal to 100. 
If it isn't the case, the phrase 'better luck next time ;)' will be printed.



Below is the result you will get when you run the program:


```python
q1= input("In which year did the World War 1 begin? ")
a1= '1914'

q2= input("What is the capital of Canada? ")
a2= 'Ottawa'

q3= input("solve for x: 2x+4 = 3x+10 ")
a3= '-6'

q4= input("Name the chemical element which's symbol is 'O' ")
a4= 'Oxygen'

q5= input("Who is the famous french author who wrote 'Les Miserables'? ")
a5= 'Victor Hugo'

res = 0

if q1 == a1:
    res = res+20

if q2 == a2:
    res = res+20

if q3 == a3:
    res = res+20
    
if q4 == a4:
    res = res+20
    
if q5 == a5:
    res = res+20

print('score:'+str(res))

if res == 100:
    print('Congratulations!')
else:
    print('Better luck next time ;)')
```
